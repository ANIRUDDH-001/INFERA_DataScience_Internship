```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import glob
```

```python
# Step 1: Get all traffic CSV files
traffic_files = glob.glob("../data/raw/dataset_2/*_Traffic.csv") 
```

```python
# Step 2: Load files with encoding fallback
traffic_dataframes = []

for file in traffic_files:
    try:
        df = pd.read_csv(file)
    except UnicodeDecodeError:
        df = pd.read_csv(file, encoding='ISO-8859-1')
    traffic_dataframes.append(df)
```

```python
# Step 3: Concatenate into one DataFrame
traffic_data = pd.concat(traffic_dataframes, ignore_index=True)
```

```python
# Step 4: Preview the loaded data
traffic_data.head()
```

```python
traffic_data.info()
```

```python
traffic_data.isnull().sum()
```

```python
traffic_data.rename(columns={'Noon': '12pm'}, inplace=True)

# 1. Drop "Total" row if it exists
traffic_data = traffic_data[~traffic_data['Sensor'].str.lower().str.contains('total', na=False)]
```

```python
# 2. Basic data information
print("Traffic Data Overview:")
print(f"Dataset shape: {traffic_data.shape}")
print(f"Number of sensors: {traffic_data['Sensor ID'].nunique()}")
print("\nFirst few sensor names:")
print(traffic_data['Sensor'].dropna().head(10).tolist())  # added dropna()
```

```python
# 3. Check for missing values
print("\nMissing values in traffic data (top 10 columns):")
print(traffic_data.isnull().sum().sort_values(ascending=False).head(10))
```

```python
# 4. Define hour columns
hour_columns = [
    'Midnight', '1am', '2am', '3am', '4am', '5am', '6am', '7am', '8am', '9am', '10am', '11am',
    '12pm', '1pm', '2pm', '3pm', '4pm', '5pm', '6pm', '7pm', '8pm', '9pm', '10pm', '11pm'
]
```

```python
# 5. Melt the data
traffic_long = pd.melt(
    traffic_data,
    id_vars=['Sensor', 'Sensor ID'],
    value_vars=hour_columns,
    var_name='Hour',
    value_name='Pedestrian_Count'
)
```

```python
# 6. Clean melted data
traffic_long['Pedestrian_Count'] = pd.to_numeric(
    traffic_long['Pedestrian_Count'], errors='coerce'
).fillna(0)
```

```python
# 7. Add time-based features
hour_mapping = {
    'Midnight': 0, '1am': 1, '2am': 2, '3am': 3, '4am': 4, '5am': 5,
    '6am': 6, '7am': 7, '8am': 8, '9am': 9, '10am': 10, '11am': 11,
    '12pm': 12, '1pm': 13, '2pm': 14, '3pm': 15, '4pm': 16, '5pm': 17,
    '6pm': 18, '7pm': 19, '8pm': 20, '9pm': 21, '10pm': 22, '11pm': 23
}
traffic_long['Hour_24'] = traffic_long['Hour'].map(hour_mapping)

def get_time_period(hour):
    if pd.isna(hour):
        return 'Unknown'
    elif 6 <= hour <= 9:
        return 'Morning Peak'
    elif 10 <= hour <= 16:
        return 'Daytime'
    elif 17 <= hour <= 19:
        return 'Evening Peak'
    elif 20 <= hour <= 23:
        return 'Evening'
    else:
        return 'Night'

traffic_long['Time_Period'] = traffic_long['Hour_24'].apply(get_time_period)
```

```python
# 8. Categorize sensor locations
def categorize_sensor_location(sensor_name):
    if pd.isna(sensor_name):
        return 'Unknown'
    sensor_lower = sensor_name.lower()
    if 'mall' in sensor_lower or 'shopping' in sensor_lower:
        return 'Retail'
    elif 'station' in sensor_lower or 'central' in sensor_lower:
        return 'Transport Hub'
    elif 'bridge' in sensor_lower or 'park' in sensor_lower:
        return 'Outdoor'
    elif 'hall' in sensor_lower or 'town hall' in sensor_lower:
        return 'Public Building'
    else:
        return 'Mixed Use'

traffic_long['Location_Type'] = traffic_long['Sensor'].apply(categorize_sensor_location)
```

```python
# 9. Summary statistics
daily_summary = traffic_long.groupby(['Sensor', 'Sensor ID', 'Location_Type']).agg(
    Daily_Total=('Pedestrian_Count', 'sum'),
    Daily_Average=('Pedestrian_Count', 'mean'),
    Daily_Peak=('Pedestrian_Count', 'max'),
    Daily_Std=('Pedestrian_Count', 'std')
).round(2).reset_index()
```

```python
# 11. Save processed data
traffic_long.to_csv('traffic_data_cleaned.csv', index=False)
daily_summary.to_csv('traffic_daily_summary.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')
```

```python
# === Load data ===

df = pd.read_csv('traffic_data_cleaned.csv')

# Feature prep 
features = pd.get_dummies(df[['Hour_24', 'Location_Type', 'Time_Period']])
target = df['Pedestrian_Count']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# === Train models ===
lr = LinearRegression().fit(X_train, y_train)
rf = RandomForestRegressor(n_estimators=100, random_state=42).fit(X_train, y_train)

lr_pred = lr.predict(X_test)
rf_pred = rf.predict(X_test)

print("\nModel Results:")
print(f"Linear RÂ²: {r2_score(y_test, lr_pred):.3f}, MAE: {mean_absolute_error(y_test, lr_pred):.0f}")
print(f"Random Forest RÂ²: {r2_score(y_test, rf_pred):.3f}, MAE: {mean_absolute_error(y_test, rf_pred):.0f}")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
# 1. Linear actual vs predicted
axes[0,0].scatter(y_test, lr_pred, alpha=0.5)
axes[0,0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
axes[0,0].set_title('Linear Regression: Actual vs Predicted')
```

```python
# 2. RF actual vs predicted
axes[0,1].scatter(y_test, rf_pred, alpha=0.5, color='darkorange')
axes[0,1].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
axes[0,1].set_title('Random Forest: Actual vs Predicted')
```

```python
# 3. Feature importance (Top 8)
importances = pd.Series(rf.feature_importances_, index=features.columns)
top_features = importances.sort_values(ascending=False).head(8)
axes[1,0].barh(top_features.index[::-1], top_features.values[::-1])
axes[1,0].set_title('Top 8 Features (Random Forest)')
```

```python
# 4. Hour-wise traffic pattern (actual vs predicted)
avg_actual = df.groupby('Hour_24')['Pedestrian_Count'].mean()
avg_pred = rf.predict(features.groupby(df['Hour_24']).mean())

axes[1,1].plot(avg_actual.index, avg_actual.values, 'o-', label='Actual')
axes[1,1].plot(avg_actual.index, avg_pred, 's-', label='Predicted')
axes[1,1].set_title('Hourly Traffic: Actual vs Predicted')
axes[1,1].legend()
```

```python
plt.tight_layout()
plt.show()

print("Best Model:", "Random Forest" if r2_score(y_test, rf_pred) > r2_score(y_test, lr_pred) else "Linear Regression")
```

```python
import joblib
import json
```

```python
# CHOOSE BEST TRAFFIC MODEL
best_traffic_model = rf if r2_score(y_test, rf_pred) > r2_score(y_test, lr_pred) else lr
best_traffic_model_name = "Random Forest" if r2_score(y_test, rf_pred) > r2_score(y_test, lr_pred) else "Linear Regression"

# Save model
joblib.dump(best_traffic_model, 'traffic_model.pkl')
```

```python
# Save traffic model info
traffic_info = {
    'features': list(features.columns),
    'model_type': best_traffic_model_name,
    'r2_score': max(r2_score(y_test, rf_pred), r2_score(y_test, lr_pred))
}

with open('traffic_model_info.json', 'w') as f:
    json.dump(traffic_info, f)
```

```python
# Sample prediction function (manual dummy input encoding will be handled in dashboard)
def predict_traffic(hour, location_type, time_period):
    """
    Simplified prediction assuming one-hot encoding like:
    Hour_24 = 14
    Location_Type = 'Retail'
    Time_Period = 'Daytime'
    """
    input_df = pd.DataFrame(columns=features.columns)
    input_df.loc[0] = 0  # set all columns to 0 initially
    input_df.at[0, 'Hour_24'] = hour

    # Set the corresponding one-hot encoded category columns to 1
    loc_col = f'Location_Type_{location_type}'
    time_col = f'Time_Period_{time_period}'

    if loc_col in input_df.columns:
        input_df.at[0, loc_col] = 1
    if time_col in input_df.columns:
        input_df.at[0, time_col] = 1

    # Predict
    return int(best_traffic_model.predict(input_df)[0])
```

```python
# RESULTS 
print("\n=== TRAFFIC MODEL EVALUATION ===")
print(f"Best Model: {best_traffic_model_name} (RÂ² = {traffic_info['r2_score']:.3f})")
print(f"Sample Prediction: {predict_traffic(14, 'Retail', 'Daytime')} people")
```
