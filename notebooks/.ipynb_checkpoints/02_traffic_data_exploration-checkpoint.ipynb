```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import glob
```

```python
# Step 1: Get all traffic CSV files
traffic_files = glob.glob("../data/raw/dataset_2/*_Traffic.csv") 
```

```python
# Step 2: Load files with encoding fallback
traffic_dataframes = []

for file in traffic_files:
    try:
        df = pd.read_csv(file)
    except UnicodeDecodeError:
        df = pd.read_csv(file, encoding='ISO-8859-1')
    traffic_dataframes.append(df)
```

```python
# Step 3: Concatenate into one DataFrame
traffic_data = pd.concat(traffic_dataframes, ignore_index=True)
```

```python
# Step 4: Preview the loaded data
traffic_data.head()
```

```python
traffic_data.info()
```

```python
traffic_data.isnull().sum()
```

```python
traffic_data.rename(columns={'Noon': '12pm'}, inplace=True)

# 1. Drop "Total" row if it exists
traffic_data = traffic_data[~traffic_data['Sensor'].str.lower().str.contains('total', na=False)]
```

```python
# 2. Basic data information
print("Traffic Data Overview:")
print(f"Dataset shape: {traffic_data.shape}")
print(f"Number of sensors: {traffic_data['Sensor ID'].nunique()}")
print("\nFirst few sensor names:")
print(traffic_data['Sensor'].dropna().head(10).tolist())  # added dropna()
```

```python
# 3. Check for missing values
print("\nMissing values in traffic data (top 10 columns):")
print(traffic_data.isnull().sum().sort_values(ascending=False).head(10))
```

```python
# 4. Define hour columns
hour_columns = [
    'Midnight', '1am', '2am', '3am', '4am', '5am', '6am', '7am', '8am', '9am', '10am', '11am',
    '12pm', '1pm', '2pm', '3pm', '4pm', '5pm', '6pm', '7pm', '8pm', '9pm', '10pm', '11pm'
]
```

```python
# 5. Melt the data
traffic_long = pd.melt(
    traffic_data,
    id_vars=['Sensor', 'Sensor ID'],
    value_vars=hour_columns,
    var_name='Hour',
    value_name='Pedestrian_Count'
)
```

```python
# 6. Clean melted data
traffic_long['Pedestrian_Count'] = pd.to_numeric(
    traffic_long['Pedestrian_Count'], errors='coerce'
).fillna(0)
```

```python
# 7. Add time-based features
hour_mapping = {
    'Midnight': 0, '1am': 1, '2am': 2, '3am': 3, '4am': 4, '5am': 5,
    '6am': 6, '7am': 7, '8am': 8, '9am': 9, '10am': 10, '11am': 11,
    '12pm': 12, '1pm': 13, '2pm': 14, '3pm': 15, '4pm': 16, '5pm': 17,
    '6pm': 18, '7pm': 19, '8pm': 20, '9pm': 21, '10pm': 22, '11pm': 23
}
traffic_long['Hour_24'] = traffic_long['Hour'].map(hour_mapping)

def get_time_period(hour):
    if pd.isna(hour):
        return 'Unknown'
    elif 6 <= hour <= 9:
        return 'Morning Peak'
    elif 10 <= hour <= 16:
        return 'Daytime'
    elif 17 <= hour <= 19:
        return 'Evening Peak'
    elif 20 <= hour <= 23:
        return 'Evening'
    else:
        return 'Night'

traffic_long['Time_Period'] = traffic_long['Hour_24'].apply(get_time_period)
```

```python
# 8. Categorize sensor locations
def categorize_sensor_location(sensor_name):
    if pd.isna(sensor_name):
        return 'Unknown'
    sensor_lower = sensor_name.lower()
    if 'mall' in sensor_lower or 'shopping' in sensor_lower:
        return 'Retail'
    elif 'station' in sensor_lower or 'central' in sensor_lower:
        return 'Transport Hub'
    elif 'bridge' in sensor_lower or 'park' in sensor_lower:
        return 'Outdoor'
    elif 'hall' in sensor_lower or 'town hall' in sensor_lower:
        return 'Public Building'
    else:
        return 'Mixed Use'

traffic_long['Location_Type'] = traffic_long['Sensor'].apply(categorize_sensor_location)
```

```python
# 9. Summary statistics
daily_summary = traffic_long.groupby(['Sensor', 'Sensor ID', 'Location_Type']).agg(
    Daily_Total=('Pedestrian_Count', 'sum'),
    Daily_Average=('Pedestrian_Count', 'mean'),
    Daily_Peak=('Pedestrian_Count', 'max'),
    Daily_Std=('Pedestrian_Count', 'std')
).round(2).reset_index()
```

```python
# 11. Save processed data
traffic_long.to_csv('traffic_data_cleaned.csv', index=False)
daily_summary.to_csv('traffic_daily_summary.csv', index=False)
```

```python
# Set plotting style
plt.style.use('default')
sns.set_palette("husl")

# Create figure and axes
fig, axes = plt.subplots(3, 2, figsize=(15, 18))
```

```python
# 1. Average Pedestrian Count by Hour
hourly_avg = traffic_long.groupby('Hour_24')['Pedestrian_Count'].mean().sort_index()
axes[0, 0].plot(hourly_avg.index, hourly_avg.values, marker='o', linewidth=2, color='blue')
axes[0, 0].set_title('Average Pedestrian Count by Hour', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Hour (24-hour format)')
axes[0, 0].set_ylabel('Average Pedestrian Count')
axes[0, 0].grid(True, alpha=0.3)
```

```python
# 2. Total Traffic by Location Type
location_traffic = traffic_long.groupby('Location_Type')['Pedestrian_Count'].sum().sort_values()
axes[0, 1].bar(location_traffic.index, location_traffic.values, color='coral', alpha=0.7)
axes[0, 1].set_title('Total Traffic by Location Type', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('Total Pedestrian Count')
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].grid(True, alpha=0.3)
```

```python
# 3. Average Traffic by Time Period
period_order = ['Night', 'Morning Peak', 'Daytime', 'Evening Peak', 'Evening']
period_traffic = traffic_long.groupby('Time_Period')['Pedestrian_Count'].mean().reindex(period_order)
axes[1, 0].bar(period_traffic.index, period_traffic.values, color='lightblue', alpha=0.7)
axes[1, 0].set_title('Average Traffic by Time Period', fontsize=14, fontweight='bold')
axes[1, 0].set_ylabel('Average Pedestrian Count')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(True, alpha=0.3)
```

```python
# 4. Top 10 Busiest Sensors
top_sensors = daily_summary.nlargest(10, 'Daily_Average')
axes[1, 1].barh(range(len(top_sensors)), top_sensors['Daily_Average'], color='gold', alpha=0.7)
axes[1, 1].set_yticks(range(len(top_sensors)))
axes[1, 1].set_yticklabels([
    name if len(name) <= 25 else name[:22] + '...' for name in top_sensors['Sensor']
])
axes[1, 1].invert_yaxis()
axes[1, 1].set_xlabel('Daily Average Pedestrian Count')
axes[1, 1].set_title('Top 10 Busiest Sensors', fontsize=14, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3)
```

```python
# 5. Pedestrian Count Distribution
axes[2, 0].hist(traffic_long['Pedestrian_Count'], bins=50, color='lightgreen', alpha=0.7)
axes[2, 0].set_title('Distribution of Pedestrian Counts', fontsize=14, fontweight='bold')
axes[2, 0].set_xlabel('Pedestrian Count')
axes[2, 0].set_ylabel('Frequency')
axes[2, 0].set_yscale('log')
axes[2, 0].grid(True, alpha=0.3)
```

```python
# 6. Heatmap: Time Period vs Location Type
pivot_data = traffic_long.groupby(['Time_Period', 'Location_Type'])['Pedestrian_Count'].mean().unstack()
if pivot_data is not None and not pivot_data.empty:
    im = axes[2, 1].imshow(pivot_data.values, cmap='YlOrRd', aspect='auto')
    axes[2, 1].set_title('Traffic Heatmap: Time Period vs Location Type', fontsize=14, fontweight='bold')
    axes[2, 1].set_xticks(range(len(pivot_data.columns)))
    axes[2, 1].set_xticklabels(pivot_data.columns, rotation=45)
    axes[2, 1].set_yticks(range(len(pivot_data.index)))
    axes[2, 1].set_yticklabels(pivot_data.index)
    plt.colorbar(im, ax=axes[2, 1], label='Average Pedestrian Count')
else:
    axes[2, 1].text(0.5, 0.5, 'No data to display', ha='center', va='center')
```
