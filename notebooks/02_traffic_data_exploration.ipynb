```python
import pandas as pd
import glob
```

```python
# Step 1: Get all traffic CSV files
traffic_files = glob.glob("../data/raw/dataset_2/*_Traffic.csv") 
```

```python
# Step 2: Load files with encoding fallback
traffic_dataframes = []

for file in traffic_files:
    try:
        df = pd.read_csv(file)
    except UnicodeDecodeError:
        df = pd.read_csv(file, encoding='ISO-8859-1')
    traffic_dataframes.append(df)
```

```python
# Step 3: Concatenate into one DataFrame
traffic_data = pd.concat(traffic_dataframes, ignore_index=True)
```
